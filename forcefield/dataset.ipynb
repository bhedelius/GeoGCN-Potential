```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from anidataset import *
```

```python
dataset = AniDataset()
```

```python
forces = dataset.forces
```

```python
tot_mags = []
ave_mags
mags = []
for force in forces:
    tot_mag = torch.norm(torch.sum(force,dim=0))
    tot_mags.append(tot_mag)
    ave_mag = torch.norm(torch.mean(force,dim=0))
    ave_mags.append(ave_mag)
    mag = list(torch.norm(force,dim=1))
    mags.extend(mag)
```

```python
plt.hist(ave_mags,bins=25,density=True)
plt.title('Total force magnitudes')
plt.ylabel('Frequency')
plt.xlabel('ANI forces')
plt.savefig('tot_force.png')
```

```python
plt.hist(ave_mags,bins=25)
plt.title('Mean force magnitudes')
plt.ylabel('Frequency')
plt.xlabel('ANI forces')
plt.savefig('mean_force.png')
```

```python
plt.hist(mags, bins=25)
plt.title('Individual force magnitudes')
plt.ylabel('Frequency')
plt.xlabel('ANI forces')
plt.savefig('indiv_force.png')
```
